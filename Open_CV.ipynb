```python
# imshow function in the OpenCV library is used to build an OpenCV window
cv2.imshow('Image',img) 
cv2.waitKey()
# Two parameters are passed to the function.# 1st parameter --> Window Name
                                            # 2nd parameter --> Image
# waitkey is used to give delay to the imshow window. It also takes a key press by the user from the keyboard.
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  
# cvt color is used to convert one color format to the other color format.
# The image to be converted is passed as the first parameter and flag is the second parameter which specifies the \n
# conversion from RGB to gray,gray to RGB
                               
```

```python
cv2.imshow('Gray_Scaled_Image',gray_img)
cv2.waitKey()
```

```python
blur_img=cv2.blur(img,(30,30))  ## Kernel size represents the intensity of the blurness.
```

```python
cv2.imshow('Blurred_Image',blur_img)
cv2.waitKey()
```

```python
edge_img=cv2.Laplacian(img,cv2.CV_64F)  ## Laplace Transform is applied to detect the edges. CV_64F is default image datatype.
```

```python
cv2.imshow('Edge_Image',edge_img)
cv2.waitKey()
```

```python
cv2.imshow('Image',img) 
cv2.imshow('Gray_Scaled_Image',gray_img)
cv2.imshow('Blurred_Image',blur_img)
cv2.imshow('Edge_Image',edge_img)
cv2.waitKey()
```

```python
video=cv2.VideoCapture(r"C:\Users\Public\Videos\Sample Videos\Wildlife.wmv") 

## videocapture is a function in OpenCV used to read videos or lively capture videos.
## when 0 is passed the video is captured from a primary webcam.
## when 1 is passed the video is captured from a secondary webcam.
## Path of the video from the local pc can be given as an input.
        
```

```python
 ## A Function called read in the Video Capture reads each and every frame from the video. 
                                                                                     ## read function returns two parameters
                           ## First Parameter is a boolean. --> True - Next Frame Exists ; False - Next Frame doesn't Exist.
                           ## Second Parameter is the frame. --> The picture from the video is returned.
while True:          ## While loop is used to continuously get a frame from the video and make a video. It is an infinite loop.
    data,frame=video.read()
    cv2.imshow("Video",frame)
    cv2.imshow("Blur Image",cv2.blur(frame,(10,10)))
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
video.release()
cv2.destroyAllWindows()  ## There is a function called as destroy all the windows which closes all the windows programatically.
    
```
